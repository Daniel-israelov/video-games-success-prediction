# SteamSpy:
Creating dataframe using steam spy API.

```python
#imports
import requests
import pandas as pd
import json
```

```python
# get API and keep the important columns. 

#build first dataframe
url = "https://steamspy.com/api.php?request=all"
res = requests.get(url)
data_json = res.json()
steam_spy_all = pd.DataFrame.from_dict(data_json, orient='index')
steam_spy_all.sort_values('positive', axis=0, ascending=False)

#concat other dataframes
for i in range(1,5):
    URL = url + "&page=" + "{0}".format(i)
    res = requests.get(URL)
    data_json = res.json()
    new_pd = pd.DataFrame.from_dict(data_json, orient='index')
    steam_spy_all = pd.concat([steam_spy_all, new_pd], axis = 0)
    
#drop unnecessary columns
steam_spy_all.drop(columns=['ccu', 'discount', 'initialprice', 'median_2weeks', 'average_2weeks', 'userscore', 'score_rank', 'appid'], inplace=True)

#sort by most positive at the top
steam_spy_all = steam_spy_all.sort_values('positive', axis=0, ascending=False)
steam_spy_all
```

```python
#droping dups'
steam_spy_all.drop_duplicates(keep='first', inplace=True)
steam_spy_all
```

```python
#removing spacial editions games
steam_spy_all = steam_spy_all[steam_spy_all["name"].str.contains("Edition")==False]
steam_spy_all.info()
```

```python
steam_spy_all.to_csv("steam_spy_clean.csv")
```
